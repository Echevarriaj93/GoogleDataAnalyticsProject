Merging Dataframes

I will merge the activity data along with the sleep and weight data to find correlation there based in the fact that they are three of the largest factors that lead to a healthy lifestyle.

```r

activeweight <- merge(activity, weight, by=c('Id'))
```

```r

activesleep <- merge(activity, sleep, by=c('Id'))
```

**Stage 3 - Analysis**

Analysis

The longer the distance the more calories burned. This also shows users are using there products because a positive relationship is shown between longer distances and calories burned.

```r


ggplot(data = activity) + geom_point(mapping = aes(x = TotalDistance  , y = Calories) , color = 'blue') 
```

Highly Active users burned the most calories in less minutes. Light Active users burned calories for the most prolonged period of time.

```r

ggplot(data = activity) + geom_smooth(mapping = aes(x = VeryActiveMinutes  , y = Calories ) , color = 'blue') +  geom_smooth(mapping = aes(x = FairlyActiveMinutes  , y = Calories ) , color = 'Orange') + geom_smooth(mapping = aes(x = LightlyActiveMinutes  , y = Calories ) , color = 'yellow') + labs(x = "Highly Active (Blue) , Fairly Active (Orange) , Lightly Active (Yellow) ", y = "Calories")
```

The lighter in weight the more active or perhaps even the more steps taken the lighter one became and the more active as well. The most active individuals weighed between 150 and 200 pounds.

```r

ggplot(data = activeweight) + geom_smooth(mapping = aes(x = WeightPounds, y = TotalSteps ))
```

People who are more active sleep better. We see more sleep in relation to fewer calories burned up until the 3000 when sleep increases likely as a result of increased activity. This hints that being more sedentary leads to more overall sleep but more activity leads to needing more sleep for recovery. The latter likely being a healthier balance at around 8-9 hours of sleep compared to the sedentary 12 hours of sleep.

```r

ggplot(data = activesleep) + geom_smooth(mapping = aes (x = Calories, y = TotalTimeInBed))
```

```r

ggplot(data = activesleep)+ geom_smooth(mapping = aes(x= SedentaryMinutes, y = TotalMinutesAsleep ))
```

Summary

##1.There is a positive relationship between activity and calories burned in the user data meaning that the users are engaged with the product.

##2.  More intense activity leads to more calories burned in less time.

##3.There seems to be a dual-relationship between being sedentary and active when it comes to sleep. Sedentary users sleep more likely due to being used to a sedentary life-style. Users who are moderately active sleep a healthy amount but less than those that are sedentary. Active users also see higher durations of sleep likely due to needing recovery.

##4.  Heavier individuals are less active. Individuals between 150-200 pounds are the most active.
